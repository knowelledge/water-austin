```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
from pandas.plotting import scatter_matrix
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
import seaborn
seaborn.set(style='ticks')
plt.style.use('bmh')
```

# Austin Water Data

###### Residential Consumption

```python
df_residential = pd.read_csv('/Users/nowellk/projects/water-austin/data/Austin_Water_-_Residential_Water_Consumption.csv')
df_residential = df_residential[df_residential['Year Month'] != 204406] # remove outlier
df_residential = df_residential[df_residential['Year Month'] != 201710] #remove partial consumption data for Oct 2017
df_residential['Year Month'] = pd.to_datetime(df_residential['Year Month'], format = '%Y%m')
```

###### Commercial Consumption

```python
df_commercial = pd.read_csv('/Users/nowellk/projects/water-austin/data/Austin_Water_-_Commercial_Water_Consumption.csv')
df_commercial['Year Month'] = pd.to_datetime(df_commercial['Year Month'], format = '%Y%m')
```

# Residential Trend Plots by Zip Code

```python
'''
Plot Residential Trend by Zip Code
'''
fg = seaborn.FacetGrid(data=df_residential, col='Postal Code', col_wrap=2, hue='Customer Class', hue_order=np.unique(df_residential['Customer Class']), aspect=3)
_ = fg.map(plt.scatter, 'Year Month', 'Total Gallons', edgecolor='w').add_legend()
```

# Commercial Trend Plots by Zip Code

```python
'''
Plot Commercial Trend by Zip Code
'''
fg = seaborn.FacetGrid(data=df_commercial, col='Postal Code', col_wrap=2, hue='Customer Class', hue_order=np.unique(df_commercial['Customer Class']), aspect=3)
_ = fg.map(plt.scatter, 'Year Month', 'Total Gallons', edgecolor='w').add_legend()
```
